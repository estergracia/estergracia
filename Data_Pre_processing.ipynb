## Load Dataset (from GDrive)

```python
import gdown
import pandas as pd

# Link : https://drive.google.com/file/d/1cSGaPGRM5wtEsn1fANYwbzLcJyIDOFsa/view?usp=sharing

# Google Drive file ID (ambil dari URL berbagi Google Drive)
file_id = '1cSGaPGRM5wtEsn1fANYwbzLcJyIDOFsa'
url = f'https://drive.google.com/uc?id={file_id}'

# Unduh file dari Google Drive
output = 'GarudaIndonesia_TweetHarvest.csv'
gdown.download(url, output, quiet=False)

# Load dataset
df = pd.read_csv(output)

# Melihat sekilas data
print(df.head())
```

## Data Pre-processing

### Ubah tipe data

```python
df['favorite_count'] = pd.to_numeric(df['favorite_count'], errors='coerce')
df['quote_count'] = pd.to_numeric(df['quote_count'], errors='coerce')
df['reply_count'] = pd.to_numeric(df['reply_count'], errors='coerce')
df['retweet_count'] = pd.to_numeric(df['retweet_count'], errors='coerce')
```

```python
# Mengisi nilai NaN dengan 0
df['favorite_count'].fillna(0, inplace=True)
df['quote_count'].fillna(0, inplace=True)
df['reply_count'].fillna(0, inplace=True)
df['retweet_count'].fillna(0, inplace=True)
```

```python
df['favorite_count'] = df['favorite_count'].astype('int64')
df['quote_count'] = df['quote_count'].astype('int64')
df['reply_count'] = df['reply_count'].astype('int64')
df['retweet_count'] = df['retweet_count'].astype('int64')
```
