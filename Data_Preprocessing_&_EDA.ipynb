```python
#mencari data outlier menggunakan IQR
#kita akan mengimport library baru yaitu numpy
#kenapa numpy? karena library ini berfungsi aljabar linear, array, dsb

import numpy as np
df
```

```python
#mencari kuartil 1(0,25) dan kuartil 3(0,75) dari dataset
#kita menggunakan .quentile dari numpy
#kemudian menghitung IQR

q1=df.quantile(0.25)
q3=df.quantile(0.75)
IQR = q3 - q1
```

```python
#Hitung IQR minimum dan maksimum
#kalikan IQR dengan 1.5 (konstanta)


min_IQR = q1 - 1.5 * IQR
max_IQR = q3 + 1.5 * IQR

print('IQR minimum = ', min_IQR)
print('IQR maksimum = ', max_IQR)
```

```python
#Cari nilai minimun dan maksimum dari data

nilai_min = np.min(df)
nilai_max = np.max(df)

print('Nilai minimum data = ', nilai_min)
print('Nilai maksimum data = ', nilai_max)
```

kondisi di mana jika nilai minimum data lebih kecil dari IQR minimum, berarti ada low outlier, dan jika nilai maksimum data lebih besar dari IQR maksimum, berarti ada high outlier

jika kita melihat data secara manual, karena isi dataset adalah persentase maka data tidak ada yang minus dan diatas 100%. jadi semua isi dataset adalah data yang masuk akal

## Distribusi Data tiap kolom

```python
#tapi sakerang kita menambahkan library scipy.stats
#karena kita menggunakan library fitter yang berbasis scipy
#kenapa fitter? karena library ini termasuk salah satu cara terbaik
#untuk melakukan tes distribusi data

import scipy.stats as sp
df
```

### **Perbaikan**

Jelaskan tes yang digunakan oleh fitter!
dalam library fitter memiliki 4 tes distribusi yaitu:
1.   distribusi normal: distribusi yang disebut distribusi gauss dan paling banyak digunakan, distribusi ini digunakan untuk untuk menghindari terjadinya penilaian yang condong atau biasa di satu kategori saja
2.   distribusi chi-square: distribusi probabilitas teoritis yang asimetrik dan kontinu, distribusi ini digunakan untuk mengetahui perbedaan observasi dengan nilai harapan dari kelompok sampel
3.   distribusi F: distribusi probabilitas kontinu, distribusi ini digunakan untuk membandingkan variabilitas dari dua dataset yang terdistribusi secara normal
4.   distribusi t: distribusi sampel dari nilai rata-rata untuk sampel dari populasi yang terdistribusi normal. jumlah sampel yang digunakan biasanya kecil yaitu n<30

Disini saya akan mengganti library fitter menjadi library scipy.stats

jelaskan kenapa kamu memakai scipy.stats, dan kenapa kamu memilih menggunakan tes itu!? Saya menggunakan scipy.stats karena untuk melihat hasil normality test dengan menggunakan tes Shapiro-Wilk. Saya memilih tes ini karena dataset yang saya punya hanya berisi 33 sampel dan hanya tes ini yang cocok untuk melakukan nilai probabilitas (pValue) untuk menentukan apakah kolom tersebut termasuk distribusi normal atau tidak.

```python
#uji tes untuk kolom "Tahun 2009: Megawati-Prabowo (%)"

pVals = pd.DataFrame(index = ['Statistical Value/Score', 'p-Value'])
pVals['Shapiro-Wilk'] = sp.shapiro(df['Tahun 2009: Megawati-Prabowo (%)'])
pVals
```

```python
#uji tes untuk kolom "Tahun 2009: SBY-Boediono (%)"

pVals = pd.DataFrame(index = ['Statistical Value/Score', 'p-Value'])
pVals['Shapiro-Wilk'] = sp.shapiro(df['Tahun 2009: SBY-Boediono (%)'])
pVals
```

```python
#uji tes untuk kolom "Tahun 2009: Jusuf Kalla-Wiranto (%)"

pVals = pd.DataFrame(index = ['Statistical Value/Score', 'p-Value'])
pVals['Shapiro-Wilk'] = sp.shapiro(df['Tahun 2009: Jusuf Kalla-Wiranto (%)'])
pVals
```

```python
#uji tes untuk kolom "Tahun 2014: Prabowo-Hatta (%)"

pVals = pd.DataFrame(index = ['Statistical Value/Score', 'p-Value'])
pVals['Shapiro-Wilk'] = sp.shapiro(df['Tahun 2014: Prabowo-Hatta (%)'])
pVals
```

```python
#uji tes untuk kolom "Tahun 2014: Jokowi-Jusuf Kalla (%)"

pVals = pd.DataFrame(index = ['Statistical Value/Score', 'p-Value'])
pVals['Shapiro-Wilk'] = sp.shapiro(df['Tahun 2014: Jokowi-Jusuf Kalla (%)'])
pVals
```

Kesimpulan: Untuk Tes Shapiro Wilk memiliki nilai Probabilitas(p-Value) = 0,05 (5%). Jika p > 0.05, dapat kita simpulkan kolom tersebut termasuk data berdistribusi normal. Kolom yang **termasuk** dengan distribusi normal adalah 'Tahun 2009: Megawati-Prabowo (%)', 'Tahun 2009: SBY-Boediono (%)', 'Tahun 2014: Prabowo-Hatta (%)', 'Tahun 2014: Jokowi-Jusuf Kalla (%)', dan yang **tidak termasuk** distribusi normal adalah 'Tahun 2009: Jusuf Kalla-Wiranto (%)'
