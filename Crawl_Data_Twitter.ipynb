```python
import pandas as pd

# Membaca file CSV pertama
file1 = pd.read_csv('tweets-data/garudaindonesia.csv')

# Membaca file CSV kedua
file2 = pd.read_csv('tweets-data/garudaindonesia2.csv')

# Menggabungkan kedua DataFrame
combined = pd.concat([file1, file2], ignore_index=True)

# Menyimpan hasil gabungan ke file baru
combined.to_csv('tweets-data/garudaindonesiajan20-mei20.csv', index=False)
```
